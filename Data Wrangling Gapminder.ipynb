# Data Wrangling Gapminder

```python
#Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
df_co2 = pd.read_csv('co2_emissions_tonnes_per_person.csv')
df_co2.head()
```

```python
df_pop_density = pd.read_csv('population_density_per_square_km.csv')
df_pop_density.head()
```

```python
df_hdi = pd.read_csv('hdi_human_development_index.csv')
df_hdi.head()
```

```python
df_gdp = pd.read_csv('income_per_person_gdppercapita_ppp_inflation_adjusted.csv')
df_gdp.head()
```

```python
df_male_edu_25 = pd.read_csv('mean_years_in_school_men_25_to_34_years.csv')
df_male_edu_25.head()
```

```python
df_female_edu_25 = pd.read_csv('mean_years_in_school_women_25_to_34_years.csv')
df_female_edu_25.head()
```

## Change structure of dataframes with 'melt' function

```python
#The subscript 'c' denotes clean dataframe
df_co2_c = pd.melt(df_co2,
                     ['country'],
                     var_name = 'year', 
                     value_name = 'co2_emissions').sort_values(by = ['country', 'year']).reset_index(drop = True)
df_co2_c.head()
```

```python
df_pop_density_c = pd.melt(df_pop_density,
                     ['country'],
                     var_name = 'year', 
                     value_name = 'pop_density').sort_values(by = ['country', 'year']).reset_index(drop = True)
df_pop_density_c.head()
```

```python
df_hdi_c = pd.melt(df_hdi,
                     ['country'],
                     var_name = 'year', 
                     value_name = 'hdi').sort_values(by = ['country', 'year']).reset_index(drop = True)
df_hdi_c.head()
```

```python
df_gdp_c = pd.melt(df_gdp,
                     ['country'],
                     var_name = 'year', 
                     value_name = 'gdp/capita').sort_values(by = ['country', 'year']).reset_index(drop = True)
df_gdp_c.head()
```

```python
df_male_edu_25_c = pd.melt(df_male_edu_25,
                     ['country'],
                     var_name = 'year', 
                     value_name = 'edu_male_25').sort_values(by = ['country', 'year']).reset_index(drop = True)
df_male_edu_25_c.head()
```

```python
df_female_edu_25_c = pd.melt(df_female_edu_25,
                     ['country'],
                     var_name = 'year', 
                     value_name = 'edu_female_25').sort_values(by = ['country', 'year']).reset_index(drop = True)
df_female_edu_25_c.head()
```

```python
# Creation of a general dataframe using inner joins to restrict the time to the largest common time range

join1 = pd.merge(pd.merge(df_co2_c, df_pop_density_c, how = 'inner', on = ['country', 'year']), df_hdi_c, how = 'inner', on = ['country', 'year'])
join2 = pd.merge(pd.merge(join1, df_gdp_c, how = 'inner', on = ['country', 'year']), df_male_edu_25_c, how = 'inner', on = ['country', 'year'])
gen = pd.merge(join2, df_female_edu_25_c, how = 'inner', on = ['country', 'year'])

gen.head()
```

```python
# New range of time

gen.year.min(), gen.year.max()
```
